# Lectura de datos
Igual que en apartado anterior extraemos los datos del fichero train.csv. Tambien extraemos la popularidad individualmente para su uso como target.

```python
import pandas as pd
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

X = pd.read_csv("train_ap1_mcp_23_24_train.csv")
y = np.array(X["popularity"])

del X["song_name"]
del X["popularity"]
```

Dadas las multiples variables del dataset es necesario reducir la dimensionalidad para poder representar los datos completos en funcion a la popularidad. Sabemos dado el analisis de regresion que los datos no presentan una fuerte correlación. Para realizar una primera representación utilizamos PCA:

```python
from sklearn.decomposition import PCA

X_low = PCA(n_components=1).fit_transform(X)

plt.figure(figsize=(10,6))
plt.xlabel('Datos con dimensionalidad reducida')
plt.ylabel('Popularidad')
plt.title('Proyección PCA')
plt.scatter(X_low, y)
plt.show()
```

Reduciendo los datos a una dimension y utilizando popularidad en el eje "y" visulizamos el dataset en dos dimensiones. El grafico es representativo de la correlación que existe con popularidad y la tendencia de la linea de regresión para predecir popularidad con valores individuales de ser practicamente paralela al eje x.
